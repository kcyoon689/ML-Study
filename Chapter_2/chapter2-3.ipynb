##### [예제 2-11]의 테일러 급수에서 delta = 0.2, 0.1, 0.01, 0.001 일 때 계산하고, delta가 정확도에 미치는 영향을 설명하시오.
$$f(x) = \frac{1}{2}x^2$$

Taylor
$$f(x + \Delta) \simeq f(x)+f'(x)\Delta$$

```python
# Q16.Taylor series

def fx(x):
    return 1/2 * x**2

def derivative_fx(x):
    return x

def taylor(x):
    return fx(x0) + derivative_fx(x0) * (x - x0)

delta = [0.2, 0.1, 0.01, 0.001]

# print("x0 = 1")

for i in delta:
    x0 = 1
    print("delta = ", i)
    print("x = ", x0 + i)
    print("f(x) = ", fx(x0 + i))
    print("taylor = ", taylor(x0 + i))
    print("error = ", abs(fx(x0 + i) - taylor(x0 + i)))
    print("=====")
```

#### Q17.
##### [알고리즘 2-5]를 그 아래 코드로 바꾸어 쓰면, 한 번도 뽑히지 않는 샘플이 있을 수 있다. 그래도 문제가 없나? n개의 샘플이 있고, 뽑는 일을 T번 반복했을 때, 한번도 뽑히지 않은 샘플이 k개일 확률을 쓰시오.

```python
import itertools
from math import factorial
import numpy as np
import plotly.graph_objects as go

def bin_dist(n, k, t):
    ncr = factorial(n) / (factorial(k) * factorial(n - k))
    pd = ncr * ((n-k)/n)**t
    return pd

case = 5 # 샘플 수 
iter_num = 5 # 시행 횟수
k = 5 # 안뽑히는 개수

fig = go.Figure()

for j in range(k+1):
    pd = np.array(bin_dist(case, j, iter_num)) 
    # print(pd)

    fig.add_trace(
        go.Bar(
            x = [j],
            y = [pd],
            name = f'k = {j}',
        )
    )

fig.show()
```
