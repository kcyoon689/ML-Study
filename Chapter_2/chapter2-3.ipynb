##### (1) 최솟값과 최소점을 분석적으로 구하시오.

```python
import numpy as np
import plotly.graph_objects as go

p = 0.1

def fx(x):
    f = 2*x[0]**2 + 3*x[0]*x[1] + 2*x[1]**2 - 4*x[0] + 2*x[1] + 24
    return f

def gradient(x):
    a = 4*x[0] + 3*x[1] - 4
    b = 3*x[0] + 4*x[1] + 2
    return a, b

x_0 = np.array([1.0, 0.9])
gradient_x0 = np.array(gradient(x_0))

x_list = [x_0]
y_list = []
fig = go.Figure()

for i in range(1000):
    grad = np.array(gradient(x_list[i]))
    x_list.append(x_list[i] - p*grad)
    y_list.append(fx(x_list[i]))

    fig.add_trace(
        go.Scatter(
            # x=i, 
            y=y_list, 
            mode='lines+markers')
        )

    #ealry stopping
    if i > 0 and (y_list[i] - y_list[i-1] == 0):
        print("early stopping")
        print("step: {}".format(i))
        break

fig.show()

for i in range(4):
    print("Step: {i}".format(i=i))
    print("x1: {}".format(x_list[i][0]))
    print("x2: {}".format(x_list[i][1]))
    print("f(x1, x2): {}".format(fx(x_list[i])))
```

$$f(x) = \frac{1}{2}x^2$$

Taylor
$$f(x + \Delta) \simeq f(x)+f'(x)\Delta$$

```python
# Q16.Taylor series

def fx(x):
    return 1/2 * x**2

def derivative_fx(x):
    return x

def taylor(x):
    return fx(x0) + derivative_fx(x0) * (x - x0)

delta = [0.2, 0.1, 0.01, 0.001]

# print("x0 = 1")

for i in delta:
    x0 = 1
    print("delta = ", i)
    print("x = ", x0 + i)
    print("f(x) = ", fx(x0 + i))
    print("taylor = ", taylor(x0 + i))
    print("error = ", abs(fx(x0 + i) - taylor(x0 + i)))
    print("=====")
```

#### Q17.
##### [알고리즘 2-5]를 그 아래 코드로 바꾸어 쓰면, 한 번도 뽑히지 않는 샘플이 있을 수 있다. 그래도 문제가 없나? n개의 샘플이 있고, 뽑는 일을 T번 반복했을 때, 한번도 뽑히지 않은 샘플이 k개일 확률을 쓰시오.

```python
from math import factorial
import numpy as np
import plotly.graph_objects as go

def bin_dist(n, k, t):
    ncr = factorial(n) / (factorial(k) * factorial(n - k))
    pd = ncr * ((n-k)/n)**t
    return pd

case = 5 # 샘플 수 
iter_num = 5 # 시행 횟수
k = 5 # 안뽑히는 개수

fig = go.Figure()

for j in range(k+1):
    pd = np.array(bin_dist(case, j, iter_num)) 
    # print(pd)

    fig.add_trace(
        go.Bar(
            x = [j],
            y = [pd],
            name = f'k = {j}',
            marker_line_color='rgb(8,48,107)',
            marker_line_width=1)
    )

fig.show()
```
