```python
import numpy as np

def cal_weight(weight, x, bias):
    return np.dot(weight, x.T) + bias

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([1, 0]).T

weight = np.array([
                    [[1.0, -1.0], [1.2, -1.1]],
                    [[1.0, 0.5], [-1.0, 1.0]],
                    [[-0.8, 0.3], [1.0, 0.4]],
                    [[0.1, 1.3], [-0.2, -0.4]],
                    ])

bias = np.array([
                [-0.3, 1.6],
                [1.0, 0.7],
                [0.5, -0.1],
                [1.0, -0.2]
                ])
```

```python
for i in range(0, len(weight)):
    x = sigmoid(cal_weight(x, weight[i], bias[i]))
    print(x)
```
